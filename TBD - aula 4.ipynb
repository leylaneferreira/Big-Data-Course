Preparação dos dados

```python
# removendo colunas desnecessárias
data = data.drop(['Engine Fuel Type', 'Market Category', 'Vehicle Style', 'Popularity'], axis=1)
data.head(5)
```

```python
# renomeando colunas
data = data.rename(columns={'Engine HP': 'HP', 'Engine Cylinders': 'Cylinders', 'Transmission Type': 'Transmission', 'Driven_Wheels': 'Drive Mode',
                            'highway MPG': 'MPG-H', 'city mpg': 'MPG-C', 'MSRP': 'Price' })
data.head(5)
```

```python
# descobrindo linhas duplicadas

print(data.duplicated()) 

print('---------------------')

data_linhas_duplicadas = data[data.duplicated()]
print('Numero de linhas duplicadas', data_linhas_duplicadas.shape)
print(data_linhas_duplicadas)
```

```python
data.shape
```

```python
# removendo linhas duplicadas
print(data.shape)
data = data.drop_duplicates(keep='last') # parâmetros: 'first', 'last', False
print(data.shape)
```

```python
# encontrando valores nulos nos dados

print(data.isnull())
print(data.isnull().sum()) #checando valores null por variável 
print('---------------------')

# # Substituindo/removendo valores nulos
data = data.dropna(axis=0) #removendo as linhas com valores nulos
print(data.isnull().sum())
print(data.shape)

# data = data.fillna(0) # Alternativa: substituindo todos os valores nulos por 0
```
